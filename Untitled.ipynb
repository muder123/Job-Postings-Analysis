### Web scrape

```python
df_list={'title':[], 'loc_sal':[], 'body':[]}

for num in range(1, 151):
    url = 'https://www.seek.com.au/data-scientist-OR-data-analyst-OR-data-engineer-jobs?page=' + str(num)
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'lxml')
    job_links = soup.find_all('a', attrs={'class':'_2iNL7wI'})

    for i in range(len(job_links)):
        loc = []
        job = 'https://www.seek.com.au' + job_links[i].get('href')
        res = requests.get(job)
        soup = BeautifulSoup(res.content, 'lxml')

#         df_list['company'].append(soup.find('span', attrs={'class':'_3FrNV7v _2QG7TNq E6m4BZb'}).text)

        df_list['title'].append(soup.find('h1').text)

        for i in range(len(soup.find_all('dd'))):
            loc.append(soup.find_all('dd')[i].text)

        df_list['loc_sal'].append(loc)

        df_list['body'].append(soup.find_all('div', attrs={'class':"_2e4Pi2B"})[0].text)
    
```

```python
df = pd.DataFrame(df_list)
```

#### Preprocessing

```python
job_cat = df[['title', 'body']]
```

```python
scientist_index=[]
for i in job_cat.index:
    if 'data' in df.title[i].lower():
        if 'scientist' in df.title[i].lower():
            scientist_index.append(i)
        elif 'science' in df.title[i].lower():
            scientist_index.append(i)

anaylst_index=[]
for i in job_cat.index:
    if 'data' in df.title[i].lower():
        if 'analyst' in df.title[i].lower():
            anaylst_index.append(i)
        elif 'analytics' in df.title[i].lower():
            anaylst_index.append(i)
```

```python
scientist = job_cat.iloc[scientist_index, :]
analyst = job_cat.iloc[anaylst_index, :]
```

```python
scientist.title = 1
analyst.title = 0
```

```python
job_cat = scientist.append(analyst).sort_index().reset_index(drop=True)
```

```python
def wordcloud(title):
    descriptions = ''
    for i in job_cat[job_cat['title'] == title]['body'].index:
        descriptions += job_cat['body'][i]
        
    mask = np.array(Image.open("star.png"))    
    wordcloud = WordCloud(max_words=40, background_color="white", mask=mask).generate(descriptions)
    plt.figure(figsize=(10, 10))
    # Display the generated image:
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```
